```python
import pandas as pd
import utils
```

```python
pairname = "ETH_USD"
granularity = "1m"

cols = [
    "open_time","open","high","low","close","volume",
    "close_time","quote_volume","n_trades","taker_buy_base","taker_buy_quote","ignore"
]


df = pd.read_csv(utils.get_his_data_filename(pairname, granularity), header=None, names=cols)
#print(df.head(3))
```

```python
df["open_time"] = pd.to_datetime(df["open_time"], utc=True)
df["close_time"] = pd.to_datetime(df["close_time"], utc=True)

num_cols = ["open","high","low","close","volume","quote_volume","n_trades","taker_buy_base","taker_buy_quote"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="raise")
df = df.drop(columns=["ignore"])
#df.head()
```
